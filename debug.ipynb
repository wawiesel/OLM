```python
j=41114 #transition
ne=0 #enrichment axis
nm=1 #moderator density axis
nb=2 #burnup axis
print(*archive.axes_values)
y=archive.coeff[...,:,j]
yin=y #/archive.coeff[...,0:1,j] #normalize to t=0
print(np.shape(yin))
ib=None
for ie in range(archive.axes_shape[ne]):
    for im in range(archive.axes_shape[nm]):
        plt.plot(archive.axes_values[nb],yin[ie,im,:],marker='.')
plt.title('coefficient vs. burnup')
plt.show()

ie=None
for ib in range(archive.axes_shape[nb]):
    for im in range(archive.axes_shape[nm]):
        plt.plot(archive.axes_values[ne],yin[:,im,ib],marker='.')
plt.ylabel('I-127 fission yield [barns]')
plt.xlabel('enrichment [U-235 wt.%]')
plt.show()

e_list=list()
for ie in range(archive.axes_shape[ne]):
    e = archive.axes_values[ne][ie]
    c='-'
    plt.plot(archive.axes_values[nb],y[ie,0,:],c,marker='.')
    
    e_list.append(e)
plt.legend(e_list)
plt.ylabel('I-127 fission yield [barns]')
plt.xlabel('burnup [MWd]')
plt.show()

bu_list=list()
for ib in range(archive.axes_shape[nb]):
    bu = archive.axes_values[nb][ib]
    c='-'
    plt.plot(archive.axes_values[ne],y[:,0,ib],c,marker='.')
    
    bu_list.append(bu)
plt.legend(bu_list)
plt.ylabel('I-127 fission yield [barns]')
plt.show()
```

```python
rel_axes=list()
for x_list in archive.axes_values:
    dx=x_list[-1]-x_list[0]
    x0 = x_list[0]
    z=list()
    for x in x_list:
        z.append((x-x0)/dx)
    rel_axes.append(z)
print(rel_axes)
```

```python
yp=np.asarray(np.gradient(yin,*rel_axes))
print(np.shape(yp))

#gradients
#burnup plots
ib=None
for ie in range(archive.axes_shape[ne]):
    for im in range(archive.axes_shape[nm]):
        #print(archive.axes_values[ne][ie],archive.axes_values[nm][im],yp[0,ie,im,:])
        plt.plot(archive.axes_values[nb],yp[0,ie,im,:])
plt.title('gradients vs. burnup')
plt.show()

#enrichment
ie=None
for im in range(archive.axes_shape[nm]):
    for ib in range(archive.axes_shape[nb]):
        plt.plot(archive.axes_values[ne],yp[0,:,im,ib])
plt.title('gradients vs. enrichment')
plt.show()
```

```python
rel_axes=list()
for x_list in archive.axes_values:
    dx=x_list[-1]-x_list[0]
    x0 = x_list[0]
    z=list()
    for x in x_list:
        z.append((x-x0)/dx)
    rel_axes.append(z)
print(rel_axes)

deriv=list()
rhist=list()
ahist=list()
khist=list()
for k in range(archive.ncoeff):
    y=archive.coeff[...,k]
    yp=np.asarray(np.gradient(y,*rel_axes))
    n=len(archive.axes_shape)
    pack=[None]*n
    for i in range(n):
        max0 = np.amax( yp[i,...] )
        min0 = np.amin( yp[i,...] )
        ypr = yp[i,...]/(y+eps)
        rmax0 = np.amax( ypr )
        rmin0 = np.amin( ypr )
        diff_list=list()
        for j in range(n):
            ydr = np.absolute(np.diff(ypr,axis=j))
            yd = np.absolute(np.diff(yp[i,...],axis=j))
            vr = np.amax(ydr)
            diff_list.append( vr )
            rhist.append( vr )
            va = np.amax(yd)
            ahist.append( va )
            khist.append( k )
        pack[i]={'min':min0,'max':max0,'rmin':rmin0,'rmax':rmax0, 'diff': diff_list}
    deriv.append(pack)
ahist = np.asarray(ahist)
rhist = np.asarray(rhist)
khist = np.asarray(khist)
```
